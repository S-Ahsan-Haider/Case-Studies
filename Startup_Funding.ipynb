**Hypothesis Testing**

```python
# Hypothesis - 1 : The distribution of startups in countries across founded years

from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(h1['country_code'], h1['founded_year'])
chi2, pval, degrees_of_freedom, expected_counts = chi2_contingency(contingency_table)

alpha = 0.05

print('\n\nHypothesis : 1\n\n')
print("Chi-square statistic:", chi2)
print("p-value:", pval)
if pval < alpha:
    print("Reject null hypothesis: The distribution of startups in countries across founded years is statistically different.")
else:
    print("Fail to reject null hypothesis: There is no evidence to conclude that the distribution of startups in countries across founded years is different.")


# Hypothesis - 2: ANOVA - Do the no. of funding rounds affect Total Funding of startups?

h1 = df[df['country_code'] == 'USA']

h1['funding_rounds_bins'] = pd.cut(h1['funding_rounds'], bins=5, labels=['1-3', '4-6', '7-9', '10-12', '13-15'])

import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('funding_total_usd ~ funding_rounds_bins', data=h1).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

p_value = anova_table['PR(>F)'][0]

alpha = 0.05
print('\n\nHypothesis : 2\n\n')
if p_value < alpha:
    print("Reject the null hypothesis. No. of funding rounds affect total funding.")
else:
    print("Fail to reject the null hypothesis. No. of funding rounds do not affect total funding.")

print('\n\n\n\n')
```

**Relationship Between Funding Rounds and Total Funding**

```python
 # Selecting the Country from the list above

h1 = h[h['country_code'] == 'USA']

# Dividing funding rounds into bins

h1['funding_rounds_bins'] = pd.cut(h1['funding_rounds'], bins=5, labels=['1-3', '4-6', '7-9', '10-12', '13-15'])

# Plot

sns.set_palette("Set3")
sns.boxplot(x='funding_rounds_bins', y='funding_total_usd', data=h1, hue='funding_rounds_bins')
plt.xlabel('Funding Rounds')
plt.ylabel('Funding Total USD')
plt.title('Funding Total USD by Funding Rounds')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', which='both', axis='y')
plt.xticks(rotation=45)
plt.show()

# Anova Test

import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('funding_total_usd ~ funding_rounds_bins', data=h1).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

p_value = anova_table['PR(>F)'][0]

alpha = 0.05

if p_value < alpha:
    print("Significant difference in across Categories")
else:
    print("No significant difference across Categories")
```
